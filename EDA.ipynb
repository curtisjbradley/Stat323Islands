```python
from random import choice
houses = sum([village.get_houses() for village in api.get_villages()],[])
def get_random_person():
    house = choice(houses)
    if house._house_index == 'empty':
        return get_random_person()
    residents = house.get_residents()
    if len(residents) == 0:
        return get_random_person()

    return choice(residents)

for i in range(0,100):
    print(get_random_person().request_consent())
```
