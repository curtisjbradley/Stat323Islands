```python
from main import IslandsAPI
from person import Person

api = IslandsAPI()

```

```python
from random import choice
houses = sum([village.get_houses() for village in api.get_villages()],[])
def get_random_person() -> Person:
    house = choice(houses)
    if house.is_empty():
        return get_random_person()
    residents = house.get_residents()
    if len(residents) == 0:
        return get_random_person()
    res : Person = choice(residents)

    if res.get_age() < 18 or res.get_age() > 60:
        return get_random_person()
    return choice(residents)

p_list = []
for i in range(0,4):
    while True:
        p = get_random_person()
        if p in p_list:
            continue
        print(p.get_name())
        if p.request_consent():
            while not p.toggle_contact():
                pass
            p_list.append(p)
            break
    print(i)
```
